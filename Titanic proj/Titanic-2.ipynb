### Finding Correlation

```python
# Load the data
corr_train = pd.read_csv('Data/train.csv')

# Data Preprocessing
corr_train['Age'].fillna(corr_train['Age'].median(), inplace=True)
corr_train['Embarked'].fillna(corr_train['Embarked'].mode()[0], inplace=True)
corr_train.drop('Cabin', axis=1, inplace=True)
corr_train.dropna(inplace=True)

# Convert categorical variables to numerical
label_encoder = LabelEncoder()
corr_train['Sex'] = label_encoder.fit_transform(corr_train['Sex'])
corr_train = pd.get_dummies(corr_train, columns=['Embarked'], drop_first=True)

# Check the columns to ensure there are no non-numeric columns
numeric_columns = corr_train.select_dtypes(include=[np.number]).columns
corr_train = corr_train[numeric_columns]

# Create a correlation matrix
correlation_matrix = corr_train.corr()

# Plot the correlation matrix using Seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Titanic Dataset Attributes')
plt.show()
```

```python


# Exclude non-numeric and 'PassengerId' columns
numeric_columns = corr_train.select_dtypes(include=[np.number]).columns
numeric_columns = numeric_columns.drop('PassengerId')
corr_train_numeric = corr_train[numeric_columns]

# Create a correlation matrix for only 'Survived' column
correlation_matrix_survived = corr_train_numeric.corr()['Survived'].sort_values(ascending=False)

# Plot the correlation values with 'Survived'
plt.figure(figsize=(10, 6))
sns.barplot(x=correlation_matrix_survived.index, y=correlation_matrix_survived.values)
plt.title('Correlation of Attributes with Survived')
plt.xticks(rotation=45)
plt.show()
```

## Now Performing Random Forest

```python
import numpy as np
import pandas as pd


# Load the data
train_data = pd.read_csv('Data/train.csv')
test_data = pd.read_csv('Data/test.csv')

```

```python
# Data Preprocessing
# Handling missing values
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
train_data.drop('Cabin', axis=1, inplace=True)
train_data.dropna(inplace=True)

```

```python

test_data['Age'].fillna(test_data['Age'].median(), inplace=True)
test_data['Embarked'].fillna(test_data['Embarked'].mode()[0], inplace=True)
test_data.drop('Cabin', axis=1, inplace=True)
test_data.dropna(inplace=True)

```

```python

# Convert categorical variables to numerical
label_encoder = LabelEncoder()
train_data['Sex'] = label_encoder.fit_transform(train_data['Sex'])
train_data = pd.get_dummies(train_data, columns=['Embarked'], drop_first=True)

test_data['Sex'] = label_encoder.fit_transform(test_data['Sex'])
test_data = pd.get_dummies(test_data, columns=['Embarked'], drop_first=True)


```

```python
# Feature selection for test and training data
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked_Q', 'Embarked_S']
X = train_data[features]
Y = train_data['Survived']
x_test = test_data[features]

# Split the data into training and validation sets
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=42)

```

```python
# Choose a Machine Learning Algorithm (Random Forest in this example)
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train, Y_train)


```

```python
#Testing the model
test_predictions = model.predict(x_test)
print(type(test_predictions))

merged_array = np.column_stack((test_data['PassengerId'], test_predictions))

# Now 'merged_array' contains both arrays as columns
print(merged_array)
```

```python
# Evaluate the model
Y_pred = model.predict(X_val)
accuracy = accuracy_score(Y_val, Y_pred)
print("Accuracy:", accuracy)
print(classification_report(Y_val, Y_pred))

# Testing the model
test_predictions = model.predict(x_test)
print(type(test_predictions))

# Create a DataFrame for submission
random_forest_df = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': test_predictions})
```

```python
random_forest_df
```

```python
# Count the number of predicted survivors
survivor_counts = random_forest_df['Survived'].value_counts()

# Plot the bar chart
plt.bar(survivor_counts.index, survivor_counts.values, color=['red', 'green'])
plt.xlabel('Survived Prediction')
plt.ylabel('Count')
plt.title('Count of Predicted Survivors')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
# Merge predictions with original test data
merged_data = pd.merge(test_data, random_forest_df, on='PassengerId', how='outer')

# Count the number of predicted survivors
survivor_counts = merged_data['Survived'].value_counts()

# Print the count of predicted survivors
print("Count of Predicted Survivors:")
print(survivor_counts)
```
